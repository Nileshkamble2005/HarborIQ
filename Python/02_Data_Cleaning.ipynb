```python
# Import the required libraries.

import pandas as pd
import numpy as np
```

```python
# Load all project datasets.

shipments = pd.read_csv("../Dataset/shipments.csv")
customers = pd.read_csv("../Dataset/customers.csv")
carriers = pd.read_csv("../Dataset/carriers.csv")
ports = pd.read_csv("../Dataset/ports.csv")
```

```python
# Display the first 5 records before cleaning.

shipments.head()
```

```python
# Check the data types of all columns.

shipments.dtypes
```

```python
# Check the total number of missing values in each column.

shipments.isnull().sum()
```

```python
# Check the total number of duplicate records.

shipments.duplicated().sum()
```

```python
# Remove duplicate records from the shipments dataset.

shipments = shipments.drop_duplicates()
```

```python
# Verify that duplicate records have been removed.

shipments.duplicated().sum()
```

```python
# Convert timestamp columns into datetime format.

date_format = "%d-%m-%Y %H:%M"

shipments["departure_timestamp"] = pd.to_datetime(
    shipments["departure_timestamp"],
    format=date_format
)

shipments["expected_arrival_timestamp"] = pd.to_datetime(
    shipments["expected_arrival_timestamp"],
    format=date_format
)

shipments["actual_arrival_timestamp"] = pd.to_datetime(
    shipments["actual_arrival_timestamp"],
    format=date_format,
    errors="coerce"
)
```

```python
# Verify the updated data types.

shipments.dtypes
```

```python
# Check for missing values again after datatype conversion.

shipments.isnull().sum()
```

```python
# Fill missing values in numerical columns with their median.

numeric_columns = shipments.select_dtypes(include=["int64", "float64"]).columns

shipments[numeric_columns] = shipments[numeric_columns].fillna(
    shipments[numeric_columns].median()
)
```

```python
# Fill missing values in categorical columns with 'Unknown'.

categorical_columns = shipments.select_dtypes(include=["object"]).columns

shipments[categorical_columns] = shipments[categorical_columns].fillna("Unknown")
```

```python
# Verify that all missing values have been handled.

shipments.isnull().sum()
```

```python
# Standardize text columns by removing extra spaces.

text_columns = shipments.select_dtypes(include=["object"]).columns

for column in text_columns:
    shipments[column] = shipments[column].str.strip()
```

```python
# Check the statistical summary after cleaning.

shipments.describe()
```

```python
# Save the cleaned shipments dataset.

shipments.to_csv("../Dataset/cleaned_shipments.csv", index=False)
```
