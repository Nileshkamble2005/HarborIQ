```python
# Import the required libraries.

import pandas as pd
```

```python
# Load the cleaned shipments dataset.

shipments = pd.read_csv(
    "../Dataset/cleaned_shipments.csv",
    parse_dates=[
        "departure_timestamp",
        "expected_arrival_timestamp",
        "actual_arrival_timestamp"
    ]
)
```

```python
# Q1. Which carrier handled the highest number of shipments?

carrier_shipments = (
    shipments.groupby("carrier_name")
    .size()
    .sort_values(ascending=False)
)

print(carrier_shipments)
```

```python
# Q2. Which carrier generated the highest shipment revenue?

carrier_revenue = (
    shipments.groupby("carrier_name")["shipment_value_usd"]
    .sum()
    .sort_values(ascending=False)
)

print(carrier_revenue)
```

```python
# Q3. Which cargo type generated the highest shipment revenue?

cargo_revenue = (
    shipments.groupby("cargo_type")["shipment_value_usd"]
    .sum()
    .sort_values(ascending=False)
)

print(cargo_revenue)
```

```python
# Q4. Which origin countries generated the highest shipment revenue?

country_revenue = (
    shipments.groupby("origin_country")["shipment_value_usd"]
    .sum()
    .sort_values(ascending=False)
)

print(country_revenue.head(10))
```

```python
# Q5. Which customers generated the highest shipment revenue?

customer_revenue = (
    shipments.groupby("customer_id")["shipment_value_usd"]
    .sum()
    .sort_values(ascending=False)
)

print(customer_revenue.head(10))
```

```python
# Q6. What percentage of shipments arrived late?

late_percentage = (
    shipments["late_arrival_flag"]
    .value_counts(normalize=True) * 100
)

print(late_percentage)
```

```python
# Q7. Which ports handled the highest number of shipments?

top_ports = (
    shipments.groupby("origin_port")
    .size()
    .sort_values(ascending=False)
)

print(top_ports.head(10))
```

```python
# Q8. Which shipment had the highest shipment value?

highest_value = shipments.nlargest(1, "shipment_value_usd")

print(highest_value)
```

```python
# Q9. Which carrier has the highest average delay duration?

average_delay = (
    shipments.groupby("carrier_name")["delay_duration_hours"]
    .mean()
    .sort_values(ascending=False)
)

print(average_delay)
```

```python
# Q10. Display the overall business summary.

print("========== HarborIQ Business Summary ==========")
print(f"Total Shipments : {len(shipments):,}")
print(f"Total Revenue (USD) : ${shipments['shipment_value_usd'].sum():,.2f}")
print(f"Average Shipment Value : ${shipments['shipment_value_usd'].mean():,.2f}")
print(f"Total Delay Penalty : ${shipments['total_delay_penalty_usd'].sum():,.2f}")
print(f"Average Delay Hours : {shipments['delay_duration_hours'].mean():.2f}")
```
